```python


# def remove_special_characters(text):
#     # Define a regular expression to match emojis
#     emoji_pattern = re.compile("["
#                                u"\U0001F600-\U0001F64F"  # Emojis
#                                u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
#                                u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
#                                u"\U0001F700-\U0001F77F"  # Alphabetic Presentation Forms
#                                u"\U0001F780-\U0001F7FF"  # Geometric Shapes
#                                u"\U0001F800-\U0001F8FF"  # Miscellaneous Symbols
#                                u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
#                                u"\U0001FA00-\U0001FA6F"  # Extended-A
#                                u"\U0001FA70-\U0001FAFF"  # Extended-B
#                                u"\U0001F004"             # Emoticon
#                                "]+", flags=re.UNICODE)
    
#     # Remove emojis from the text
#     text = emoji_pattern.sub(r'', text)
    
#     # Remove specific characters
#     text = text.replace("\n", '')  # Remove newline characters
#     text = text.encode('ascii', 'ignore').decode()  # Remove special characters, such as \u0026
    
#     return text

def remove_special_characters(text):
    # Define a regular expression to match emojis
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emojis
                               u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                               u"\U0001F700-\U0001F77F"  # Alphabetic Presentation Forms
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes
                               u"\U0001F800-\U0001F8FF"  # Miscellaneous Symbols
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Extended-A
                               u"\U0001FA70-\U0001FAFF"  # Extended-B
                               u"\U0001F004"             # Emoticon
                               "]+", flags=re.UNICODE)
    
    # Remove emojis from the text
    text = emoji_pattern.sub(r'', text)
    
    # Remove specific characters
    text = text.replace("\\n", '')  # Remove newline characters
    text = text.replace('\\u0026', '')  # Remove '\u0026'
    text = text.replace('@', '')
    text = text.replace('.', '')
    
    return text

```

```python
# Iterate through each row in the parent DataFrame
user_reviews_data = []

for index, row in data.iterrows():
    
    user_reviews_text = row['user_reviews']
   
    
    # Check if 'user_reviews' is not NaN and contains '{' and '}'
    if not pd.isna(user_reviews_text) and '{' in user_reviews_text and '}' in user_reviews_text:
        # Split the text using '{' and '}' as delimiters
        review_blocks = user_reviews_text.split('{')
       
        
        # Iterate through review blocks
        for block in review_blocks:
            if '}' in block:
                
                # Split the block using ',' as a delimiter and extract the 3rd element
                review = block.split('"')[13]
                review = remove_special_characters(review)
           # Create a dictionary and append it to a list
                user_reviews_data.append({'hotel_index': index, 'review': review, 'tokenized_reviews':''})
                # print(f'{index} , {review}')
                
          

user_reviews = pd.concat([user_reviews, pd.DataFrame(user_reviews_data)])
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

def preprocess_text(text_list):
    cleaned_texts = []
    
    for text in text_list:
        # Lowercase the text
        text = text.lower()

        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)

        # Tokenize the text
        tokens = word_tokenize(text)

        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if word not in stop_words]

        # Join the tokens back into a single string
        cleaned_text = ' '.join(tokens)
        cleaned_texts.append(cleaned_text)
    
    return cleaned_texts
```
